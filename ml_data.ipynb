```python
import pandas as pd
import numpy as np
```

```python
file="Machine learing data.xlsx"#path for the file
```

```python
df=pd.read_excel(file,skiprows=4)#loaded the dataset
```

```python
df
```

```python
# Cleaning the data by removing unnecessary columns  --->mostly unnamed columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
#identifying the numeric and non numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns
```

```python
#function for augmentation on the numeric data
def augment_data(data, numeric_cols, n_augments=10, noise_level=0.01):
    augmented_data = data.copy()
    for _ in range(n_augments):
        noise = np.random.normal(0, noise_level, data[numeric_cols].shape) #generates the noise
        augmented_sample = data.copy()
        augmented_sample[numeric_cols] = data[numeric_cols] + noise #adds the noise
        augmented_data = pd.concat([augmented_data, augmented_sample], ignore_index=True)#further concatenates it augmented sample and original one
    return augmented_data
```

```python
# Apply data augmentation
expanded_data = augment_data(df, numeric_cols)
```

```python
# Save the expanded dataset to a new Excel file
expanded_file_path = 'Expanded_Machine_Learning_Data_Augmented.xlsx'
expanded_data.to_excel(expanded_file_path, index=False)
```
